```python

def compare(a: int, b: int):
    if a > b:
        return f"{a} is greater than {b}"
    elif a == b:
        return f"{a} is equal to {b}"
    else:
        return f"{a} is less than {b}"
```

```python
def max_of_two(a: int, b: int):
    if a > b:
        return a
    elif a < b:
        return b
    else:
        return a 
```

```python
def fubonacii_iterative():
    n=int(input("Enter the number of terms: "))
    def fibonacci(n):
        if n<=0:
            return 0
        elif n==1:
            return 1
        else:
            return fibonacci(n-1)+fibonacci(n-2)
    for i in range(n):
        print(fibonacci(i), end=' ')
```

```python
# evalutate the function 

    
def evalutate_expressioin(expr):
    try:
        result=eval(expr)
        return result
    except ZeroDivisionError:
        print("error division by zero is not allowed")
    except SyntaxError:
        print("error invalid syntax")
    except TypeError:
        print("type error")
    except Exception as e:
        print(f"error these cannot be evaluted {e}")

if __name__ == "__main__":
    expr = input("Enter a mathematical expression: ")   
    evalutate_expressioin(expr)

# result = eval(compile('2*5', 'ram', 'eval'))
# print(result)  
```

```python
def dobEception(age):
    try:
        if age < 0:
            raise ValueError("Age cannot be negative.")
        elif age <= 30 and age > 20:
            print("young age")  
    except ValueError as ve:
        print(f"Invalid input: {ve}")
    except Exception as e:
        print(f"An error occurred: {e}")    
def exception_year():
    year = int(input("Enter a year: "))
    age=2021-year
    try:
        if age <= 30 and age > 20:
            print("young age") 
        else:
            raise dobEception(age)
    except dobEception as de:
        print(f"Invalid input: {de}")
if __name__ == "__main__":
    exception_year()
```
