```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
t = int(input())
for i in range(t):
    n=int(input())
    arr=list(map(int,input().split()))
    left=0
    right=n-1
    possible=True
    if arr[left]<=arr[right]:
        last=arr[right]
        right-=1
    else:
        last=arr[left]
        left+=1
    while left<=right:
        if arr[left]>=arr[right]:
            pick=arr[left]
            left+=1
        else:
            pick=arr[right]
            right-=1
        if pick > last:
            possible=False
            break
        last=pick
    print("Yes" if possible else "No")

            
```
